```python
import networkx as nx
from scipy.sparse.csgraph import bellman_ford
```

```python
# edge (u, v, w) means an edge from u to v with weight w
edges = [
    (0, 1, 3),
    (0, 2, 1),
    (1, 2, 7),
    (1, 3, 5),
    (2, 3, 2),
    (3, 2, -1),
    (3, 4, 1),
    (4, 1, -2),
]

# assuming A-E are indexed as 0-4
distances = [float('inf')] * 5
distances[0] = 0

for _ in range(4):   # v-1 iterations
    for u, v, w in edges:   # 'w' is the 'new' weight
        if distances[u] < float('inf'):
            distances[v] = min(distances[v], distances[u] + w)

print("Shortest distances from vertex A (0):", distances)
```

```python
# build a small directed graph with weights (same as in example)
G = nx.DiGraph()
G.add_weighted_edges_from([
    ('A', 'B', 3),
    ('A', 'C', 1),
    ('B', 'C', 7),
    ('B', 'D', 5),
    ('C', 'D', 2),
    ('D', 'C', -1),
    ('D', 'E', 1),
    ('E', 'B', -2),
])

# convert to an adjacency/weight matrix
W = nx.to_numpy_array(G, weight='weight')

# run scipy's bellman_ford on the weight matrix from source index 0
distances, predecessors = bellman_ford(W, indices=0, return_predecessors=True)

nodes = list(G.nodes())
print("nodes:", nodes)
print("distances from source (A):", distances)
print("predecessors:", predecessors)

# interpret results in terms of node names
for i, d in enumerate(distances):
    pred = predecessors[i]
    pred_name = nodes[pred] if pred != -9999 else None
    print(f"  {nodes[i]}: distance={d}, predecessor={pred_name}")
```
